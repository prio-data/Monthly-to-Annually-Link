# some other advanced/automated methods of querying the data

```python
def shorten_name(string, part_to_delete):
    output = string.replace(part_to_delete, '')
    return output

def loop_query(table_name, variable_list, qs_name, level, remove_this):
    table_wanted = table_name
    all_variable_list = variable_list
    qs = (Queryset(f"{table_wanted}_{qs_name}", level))
    
    for variable in all_variable_list:
        variable_cleaned = variable.strip()
        new_name = shorten_name(variable_cleaned, remove_this)
        qs = qs.with_column(Column(f"{new_name}",from_table = table_wanted, from_column = variable_cleaned))
    
    data = qs.publish().fetch()
    return data

def extract_all_variable(table_name):
    table_wanted = table_name
    raw = subprocess.run(['viewser', 'tables', 'show' ,table_wanted], stdout=subprocess.PIPE)
    temp = [item for item in raw.stdout.decode().split('|') if '\n' not in item][6:][::2]
    all_variable_list = []
    for i in temp:
        variable = i.strip()
        all_variable_list.append(variable)
    return all_variable_list

def query_all_table_no_fill_short(table_name, level, remove_this):
    table_wanted = table_name
    all_variable_list = extract_all_variable(table_wanted)

    data =  loop_query(table_wanted, all_variable_list, 'all', level, remove_this)
    
    return data
```
